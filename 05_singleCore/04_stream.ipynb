Measures the sustainable memory performance of the system in MB/s.

## Stream without optimization

```python
%%bash
icc --version
icc -DSTREAM_ARRAY_SIZE=100000000 -g stream.c -mcmodel=medium -o stream_noopt.exe
```

```python
!./stream_noopt.exe
```

## Stream optimized

```python
%%bash
icc -DSTREAM_ARRAY_SIZE=100000000 -mcmodel=medium -Ofast -simd -xhost -vec_report=3 stream.c -o stream.exe
```

```python
!./stream.exe
```
