# Jacobi program

\begin{equation}
\frac{\partial \Phi}{\partial t}$ = \nabla \Phi
\end{equation}

Discretization with 2nd order finite differences

\begin{equation}
\frac{\delta \Phi(x_i, y_i)}{\delta t} = \frac{\Phi(x_{i+1,i},y_i)+\Phi(x_{i-1},y_i)-2\Phi(x_i,y_i)}{(\delta x)^2} +
\frac{\Phi(x_i,y_{i+1})+\Phi(x_i,y_{i-1})-2\Phi(x_i,y_i)}{(\delta x)^2}
\end{equation}
